## 2. REALIZAR UN HISTOGRAMA CON LOS SALARIOS. QUE RANGOS DE SALARIOS SON LOS MAS POPULARES?

## 4. GRAFICAR LA SERIE DE TIEMPO CORRESPONDIENTE A LA FECHA DE CONTRATACION (DATE OF HIRE)

```python
#Convierto de tipo "Object" a tipo "Fecha"
data['DateofHire'] = pd.to_datetime(data['DateofHire'], format='%m/%d/%Y')
```

```python
#Extraigo el aÃ±o
data['DateofHire'] = data['DateofHire'].dt.year
```

```python
#Creo una serie, cuyo index es el AÃ±o, y lo ordeno de forma ascendente
data11 = pd.DataFrame(data['DateofHire'].value_counts()).sort_index()
data11
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(data11.index, data11['DateofHire'], label='Contratacion')

ax.set_xlabel('Fechas de Contratacion') 
ax.set_ylabel('Cantidad de contrataciones en esa fecha') 
ax.set_title('FECHAS CONTRATACION') 
ax.legend()
```
